### Figure level function -> `catplot`

I . Categorical Scatter Plot             Categorical Plot: Between numerical and categorical values

a) Strip Plot

```python
#axes level

sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
#figure level

sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
# jitter   values can be : true/false/numbers
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=False)
```

b) swarmplot

```python
#axes level
sns.swarmplot(data=tips,x='day',y='total_bill')
```

```python
# figure level

sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
# hue
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

II . Categorical Distribution Plots

a) Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
# axes level

sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# figure level
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
# hue
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
# single boxplot -> numerical col  || together on all values
sns.boxplot(data=tips,y='total_bill')
```

b) Violinplot = (Boxplot + KDEplot)

```python
# axes level
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# figure level
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# hue & split
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

III. Categorical Estimate Plot -> for central tendency

a) Bar Plot

```python
import numpy as np
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',ci=None)
```

b) Point Plot

```python
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',ci=None)
```

III. Regression Plots

a) regplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# axes level
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

b) Implot

```python
# axes level
# hue parameter is available
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

c)residplot

```python
#axes level
sns.residplot(data=tips,x='total_bill',y='tip')
```
